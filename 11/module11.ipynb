# Machine Learning Models Workflow

This Jupyter Notebook demonstrates the workflow for training and evaluating machine learning models using three different datasets: IMDb, Reuters, and Boston Housing.

1. Imdb dataset to learn about binary classification

```python
from imdb import IMDb
```

Create an instance of the Imdb class

```python
imdb = IMDb()
```

Prepare the data

```python
imdb.prepare_data()
```

Build the model

```python
imdb.build_model()
```

Train the model

```python
imdb.train()
```

Plot the loss

```python
imdb.plot_loss()
```

Plot accuracy

```python
imdb.plot_accuracy()
```

Evaluate the model on the test set

```python
imdb.evaluate() 
```
